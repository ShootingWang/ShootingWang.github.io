```python
import matplotlib
```

```python
import seaborn as sns   ## seaborn
sns.set()      ## seaborn
tips = sns.load_dataset("tips")  ##tips

%matplotlib inline  
##jupyter
##
fig = sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
fig.savefig('.png', dpi=400)
## dpi dots per inch
```

```python
ex = sns.load_dataset("exercise")
ex.head()

fig = sns.catplot(x="time", y="pulse", hue="kind",
                  col="diet", data=ex)
fig.savefig('strip.png', dpi=400)
```

```python
## aspect=0.8
fig = sns.catplot(x="time", y="pulse", hue="kind",
           col="diet", data=ex, aspect=0.8)
fig.savefig('-strip.png', dpi=400)
```

```python
ex = sns.load_dataset("exercise")
fig = sns.catplot(x="time", y="pulse", hue="kind",
           col="kind", data=ex)
fig.savefig('.png', dpi=400)
```

```python
##  col_wrappanel
ex = sns.load_dataset("exercise")
fig = sns.catplot(x="time", y="pulse", hue="kind",
           col="kind", data=ex, col_wrap=2)
fig.savefig('panel-.png', dpi=400)
```
